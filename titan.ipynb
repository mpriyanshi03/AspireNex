```python
import matplotlib.pyplot as plt
# Survival rate by passenger class
survival_by_class = df.groupby('Pclass')['Survived'].count()
survival_by_class.plot(kind='bar',color='blue',edgecolor='black')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
print(survival_by_class)
```

```python
#Gender Distribution 

import matplotlib.pyplot as plt
survival_by_class = df.groupby('Sex')['Survived'].count()
survival_by_class.plot(kind='bar',color='pink',edgecolor='black')
plt.title('Gerder Distribution')
plt.xlabel('Passenger Gender')
plt.ylabel('Survival Rate')
plt.show()
print(survival_by_class)
```

```python
# Survival rate by passenger class
import seaborn as sns
sns.barplot(x='Sex',y='Survived',data=df)
plt.title('Survival Rate by Gender')
plt.show()

```

```python
# Age Distribution 

plt.hist(df['Age'],bins=25,color='blue',edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Survival by Age Distribution 

sns.histplot(x='Age',hue='Survived',data=df,bins=20,kde=True)
plt.title('Age Distribution by survival status')
plt.show()
```

```python
# Select numeric columns only 
numeric_columns = df.select_dtypes(include=['float64','int64']).columns

# Correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',linewidths=0.3)
plt.title('Correlation Matrix')
plt.show()
```

```python
labels= 'Female','Male'
plt.pie(df['Sex'].value_counts(),labels=labels,autopct='%0.02f%%')
plt.legend()
plt.show()
```

```python
# Create a crosstab of Gender and Survived columns
gender_survived = pd.crosstab(df['Sex'], df['Survived'])

# Plotting the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
gender_survived.plot(kind='bar', stacked=True, ax=ax, color=['orange', 'blue'], alpha=0.75)

# Adding labels and title
ax.set_title('Total')
ax.set_xlabel('Gender')
ax.set_ylabel('Total Number')
ax.legend(["Not Survived", "Survived"], title="Survived")

# Adding counts to each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```
